```python
import pandas as pd
import requests
from io import BytesIO
import logging
from sklearn.model_selection import train_test_split
import mlflow
import matplotlib.pyplot as plt
import seaborn 
import os

current_path = os.path.dirname(os.getcwd())

file_path = current_path+"/data/05_model_input/base_test.parquet"
testData = pd.read_parquet(file_path)
file_path = current_path+"/data/05_model_input/base_train.parquet"
trainData = pd.read_parquet(file_path)


```

```python

new_labels = ['Not Shot', 'Shot']
fig, axes = plt.subplots(1, 2, figsize=(20, 10))  
seaborn.countplot(x=trainData["shot_made_flag"], ax=axes[0])
axes[0].set_xticklabels(new_labels)
axes[0].set_title("Train Data")

seaborn.countplot(x=testData["shot_made_flag"], ax=axes[1])
axes[1].set_xticklabels(new_labels)
axes[1].set_title("Test Data")

plt.tight_layout()

plt.show() 
```
